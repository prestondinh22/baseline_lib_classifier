```python
import os 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

LCO_path = "data/LCO/LCO_2_25C_Reg.csv"

LCO_all = pd.read_csv(LCO_path)
LCO_all.head()
```

```python
LCO_data = LCO_all[['Voltage(V)','Current(A)','Test_Time(s)','Cycle_Index']]
charge_mask = LCO_data[LCO_data['Current(A)'] > 0]
cycle_mask = charge_mask[charge_mask['Cycle_Index'] == 2]
time_mask = cycle_mask[cycle_mask['Test_Time(s)'] < 32513.18805]
LCO_voltage = time_mask['Voltage(V)']
LCO_time = time_mask['Test_Time(s)']
time_rel = LCO_time - LCO_time.min()
time_minutes = time_rel / 60
plt.plot(time_minutes, LCO_voltage, label="Voltage", color='blue')

plt.xlabel("Time (m)")
plt.ylabel("Voltage (V)")
plt.title("LCO Charge Over Time")

plt.legend()
plt.show()




    
```
